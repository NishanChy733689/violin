```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
var=sns.load_dataset('tips')
```

```python
var
```

```python
sns.violinplot(x=var.day,y=var.total_bill,hue=var.time)
```

```python
sns.violinplot(x=var.day,y=var.total_bill,linewidth=3,color='r')
```

```python
sns.violinplot(x=var.time,y=var.total_bill,order=['Dinner','Lunch'],inner='point')
```

```python
sns.violinplot(x=var.day,y=var.total_bill,linewidth=3,hue=var.sex,split=True,scale='width')
```

```python
sns.violinplot(var.total_bill)
```
